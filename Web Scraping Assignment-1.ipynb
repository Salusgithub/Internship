```python
import requests
from bs4 import BeautifulSoup
import pandas

topics_url = 'https://www.imdb.com/chart/top?sort=ir,desc&mode=simple&page=1'
response = requests.get(topics_url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')
results = soup.find_all("td")

lists=[]
count=1
for i in range(1,len(results),5):
    if i==len(results)-1:
        break;
    elif count>100:
        break;
    else:
        name=results[i].text.strip().split(".")[-1].strip().split("\n")[0]
        year=results[i].text.strip().split(".")[-1].strip().split("\n")[-1][1:-1]
        rating=results[i+1].text.strip()
        count=count+1
        
    lists.append([name,rating,year])

data=pandas.DataFrame(lists,columns=["Name", "Rating", "Year of Release"])

data
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

topics_url = 'https://presidentofindia.nic.in/former-presidents.htm'
response = requests.get(topics_url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')
results = soup.find_all('ul', 'listing cf')

lists=[]
for ultag in results:
    for litag in ultag.find_all('li'):
        name=litag.h3.text
        term=litag.p.text
        lists.append([name,term])

lists
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')


results=[]

data=soup.find("ul",class_="LatestNews-list").find_all("li",class_="LatestNews-item")


for i in data:
    headline=i.find("a",class_="LatestNews-headline").text
    time=i.find("time",class_="LatestNews-timestamp").text
    link=i.find("a",class_="LatestNews-headline").get('href')
    results.append([headline,time,link])
    
print(results)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
response = requests.get(url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')


results=[]

data=soup.find_all("article")


for i in data:
    title=i.find("h2").text
    authors=i.find("span",class_="sc-1w3fpd7-0 pgLAT").text
    date=i.find("span",class_="sc-1thf9ly-2 bKddwo").text
    link=i.find("a",class_="sc-5smygv-0 nrDZj").get('href')
    results.append([title,authors,date,link])
    
print(results)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

url = 'https://www.dineout.co.in/kochi'
response = requests.get(url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')


results=[]

data=soup.find("div", class_="react-multi-carousel-list").find_all("li")
print(data)

for i in data:
    name=i.find("h4").text
    #cuisine=i.find("span",class_="sc-1w3fpd7-0 pgLAT").text
    location=i.find("p",class_="_1jbOb").text.strip()
    rating=i.find("div",class_="kGUdK _1oTbl").text.strip()
    link=i.find("img").get('src')
    results.append([name,location,rating,link])
    
print(results)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

url = 'https://scholar.google.com/citations?view_op=top_venues&hl=en'
response = requests.get(url)
page_content = response.text

soup = BeautifulSoup(response.text, 'html.parser')


results=[]

data=soup.find("table", class_="gsc_mp_table").find_all("tr")
inc = 0
for i in data:
    if(inc>0):
        rank=i.find("td", class_="gsc_mvt_p").text
        pub=i.find("td", class_="gsc_mvt_t").text
        h5_index=i.find_all("td", class_="gsc_mvt_n")[0].text
        h5_median=i.find_all("td", class_="gsc_mvt_n")[1].text
        results.append([rank,pub,h5_index,h5_median])
    inc=inc+1
    
print(results)
```
